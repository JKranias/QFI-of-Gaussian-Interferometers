```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
plt.rcParams["text.usetex"]=False
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
print("hello world")
```

```python
N=0.1
rmax=np.arcsinh(np.sqrt(N))
rlist=np.linspace(0.01,rmax,1000)

#critical T to surpass MZI
def TC(N):
    if N==0.5:
        return 1/np.sqrt(2)
    
    else:
        return np.max([((2*N-1)/(8*N))*(1-np.sqrt(1+16*N/((2*N-1)**2))),((2*N-1)/(8*N))*(1+np.sqrt(1+16*N/((2*N-1)**2)))])


#range of T values
Tvals=np.linspace(0.1,0.999,1000)


#optimal seeding |a|^2 if b=0, given ropt
def optseed(r):
    return (N-np.sinh(r)**2)/(np.cosh(r)**2)

#finds M for QFI calculation
def M(s):
    l=int(len(s)/2)
    omg=np.block([[np.zeros((l,l)),np.identity(l)],[-np.identity(l),np.zeros((l,l))]])
    return np.kron(s,s)-np.kron(omg,omg)

#returns first term of QFI given s and ds
def Mtermbad(s,ds):
    return (1/2)*np.dot(ds,np.linalg.solve(M(s),ds))

def Minv(s,sinv):
    l=int(len(s)/2)
    omg=np.block([[np.zeros((l,l)),np.identity(l)],[-np.identity(l),np.zeros((l,l))]])
    evals,evecs=np.linalg.eig(np.dot(s,omg))
    q=evecs
    L=np.diag(evals)
    Linv=np.linalg.inv(np.identity(4*l**2)-np.kron(L,L)) 
    qinv=np.linalg.inv(q)
    Ainv=np.kron(sinv,sinv)
    Q=np.kron(q,q)
    Qinv=np.kron(qinv,qinv)
    return Ainv-np.linalg.multi_dot([Ainv,Q,Linv,Qinv])

#QFI of MZI
def MZI(T):  
    return 4*T*N
    
```

```python
#Internal data

#QFI for internal loss
def IY(T,r):
    return (((T*np.sinh(2*r))**2)/(1-T+T**2+T*(1-T)*np.cosh(2*r))+4*T*(1-T+T*np.cosh(2*r))*(N-np.sinh(r)**2)/(1-2*T+2*(T**2)+2*T*(1-T)*np.cosh(2*r)))


#data will be of the form [T,IQmax,r1opt]
data=[]

for i in Tvals:
    Ilist=[]
    for j in rlist:
        Ilist.append(IY(i,j))
        
    graph=[rlist,Ilist]
    IQmax=np.amax(graph,axis=1)[1]
    optcoord=np.argmax(graph,axis=1)[1]
    r1opt=rlist[optcoord]
    data.append([i,IQmax,r1opt])
    
TdataY=np.transpose(data)[0]
IdataY=np.transpose(data)[1]
rdataY=np.transpose(data)[2]
```

```python
#internal loss plots

plt.title(r"Max $I_Q$ vs. T for $N=$"+str(N)+", $\eta=1$")
plt.xlabel(r"Transmitivity $T$")
plt.ylabel(r"Max $I_Q$")
plt.plot(TdataY,IdataY,label="Mandel/Yurke")
plt.plot(TdataY,MZI(TdataY),color="black",linestyle="dashed",label="MZI")
plt.vlines(x=TC(N),ymin=np.min(IdataY),ymax=np.max(IdataY),color="red",label="$T_C=$"+str(round(TC(N),2)),linestyle="dotted")
plt.legend()
plt.show()

plt.title(r"Optimal Parameters vs. $T$ for $N=$"+str(N)+", $\eta=1$")
plt.xlabel(r"Transmitivity $T$")
plt.vlines(x=TC(N),ymin=np.min(rdataY),ymax=np.max(rdataY),color="red",label="$T_C=$"+str(round(TC(N),2)),linestyle="dotted")
plt.plot(TdataY,rdataY,label=r"Optimal $r_1$")
plt.plot(TdataY,optseed(rdataY),label=r"Optimal $|\alpha|^2$",color="black")
plt.legend()
plt.show()


Ndata=np.linspace(0.01,10,1000)
TCdataA=[]
for n in Ndata:
    TCdataA.append(TC(n))

plt.title(r"$T_C$ vs $N$ for $\eta=1$")
plt.xlabel(r"$N$")
plt.ylabel(r"$T_C$")
plt.plot(Ndata,TCdataA)
plt.show()
```

```python
#Yurke external loss

#external Yurke QFI
#p=phi-t1+t2
def EY(n,r1,r2,p):
    
    gam=np.cosh(2*r1)*np.cosh(2*r2)+np.cos(p)*np.sinh(2*r1)*np.sinh(2*r2)
    
    A=n*(np.sinh(2*r1)**2)
    
    B=2*n*gam+(1-n)*(np.cosh(2*r2)**2-np.cos(2*p)*(np.sinh(2*r2)**2))-(n+1)
    
    C=2*(gam-1)*(n*(1-n)*(gam-1)+1)
    
    D=4*n*(n*np.cosh(2*r1)+(1-n)*np.cosh(2*r2))
    
    E=N-(np.sinh(r1)**2)
    
    F=1+2*n*(1-n)*(gam-1)
    
    return A*B/C+D*E/F
```

```python
N=1

#returns optimal angle in case of high r2
def popt(N):
    num=-2*(2*N+1)*np.sqrt(N*(N+1))
    den=4*N*(N+1)+1
    return np.arccos(num/den)
    

#finds maximum QFI in case of high r2 external loss Yurke using optimal angle
def EYmax(n,r2):
    rmax=np.arcsinh(np.sqrt(N))
    return EY(n,rmax,r2,popt(N))

ndata=np.linspace(0,1,1000)


plt.title(r"Optimal $I_Q$ vs. $\eta$ of unseeded Yurke Interferometer for $N=$"+str(N)+r", $\varphi=\varphi_0$, $T=1$")
plt.xlabel(r"External Transmitivity $\eta$")
plt.ylabel(r"$I_Q$")
plt.plot(ndata,EYmax(ndata,1),label=r"$r_2=1$")
plt.plot(ndata,EYmax(ndata,3),label=r"$r_2=3$")
plt.plot(ndata,EYmax(ndata,5),label=r"$r_2=5$")
plt.hlines(4*N*(N+1),0,1,color="black",linestyle="dashed",label=r"$I_{Q_{max}}$")
plt.legend()
plt.show()
```

```python
#External Mandel with no a mode

def EMa(n,r1,r2):
    
    A=n*((np.sinh(2*r1)*np.sinh(r2))**2)/((np.cosh(r1)*np.cosh(r2))**2-1)
    B=4*n*(np.sinh(r2)**2)*(1-n+n*np.cosh(2*r1))*(N-np.sinh(r1)**2)/(1+2*n*((np.cosh(r1)*np.cosh(r2))**2-1))
    return A+B
```

```python
#optimizing for each n
N=0.1
rmax=np.arcsinh(np.sqrt(N))
r2max=5

def EMaopt(n,num,r2max):
    r1list=np.linspace(0,rmax,num)
    r2list=np.linspace(0.01,r2max,num)
    data=[]
    
    for j in r1list:
        for k in r2list:
            data.append([EMa(n,j,k),j,k])
    
    cord=np.argmax(data,axis=0)[0]
    return data[cord]

nlist=np.linspace(0,1,100)

optdata=[]

for i in nlist:
    optdata.append(EMaopt(i,100,r2max))
    
Idata=np.transpose(optdata)[0]
rdata=np.transpose(optdata)[1]
r2data=np.transpose(optdata)[2]
```

```python
plt.title(r"Optimal $I_Q$ vs. $\eta$ of Mandel Interferometer Without Mode $\hat{a}$ for $N=$"+str(N)+", $T=1$, $r_2=$"+str(r2max))
plt.xlabel(r"Transmitivity $\eta$")
plt.ylabel(r"Max $I_Q$")
plt.plot(nlist,Idata,label=r"Mandel without mode $\hat{a}$")
plt.plot(nlist,MZI(nlist),color="black",linestyle="dashed",label="MZI")
plt.legend()
plt.show()

plt.title(r"Optimal Parameters vs. $\eta$ for $N=$"+str(N)+", $\eta=1$")
plt.xlabel(r"Transmitivity $\eta$")
plt.plot(nlist,rdata,label=r"Optimal $r_1$")
plt.plot(nlist,r2data,label=r"Optimal $r_2$",color="black")
plt.legend()
plt.show()
```

```python
#QFI for Mandel with all modes

from numpy import cosh
from numpy import sinh
from numpy import cos
from numpy import sin
from numpy import sqrt


def MQFI(n,r1,r2):
    return 4*n*(N - sinh(r1)**2)*(n + (1 - n)*cosh(2*r2))*(n*cosh(2*r1) - n + 1)/(4*n*(1 - n)*(cosh(r1)**2*cosh(r2)**2 - 1) + 1) + n*(2*n*(1 - n)*(cosh(r1)**2*cosh(r2)**4 + 1) + (2*n - 1)*(n*(cosh(r1)**2 + 1) - 1)*cosh(r2)**2 - 1)*sinh(2*r1)**2/((cosh(r1)**2*cosh(r2)**2 - 1)*(2*n*(1 - n)*(cosh(r1)**2*cosh(r2)**2 - 1) + 1))
```

```python
#optimizing for each n
N=0.1
rmax=np.arcsinh(np.sqrt(N))
r2max=5

def EMopt(n,num,r2max):
    r1list=np.linspace(0.01,rmax,num)
    r2list=np.linspace(0.01,r2max,num)
    data=[]
    
    for j in r1list:
        for k in r2list:
            data.append([MQFI(n,j,k),j,k])
    
    cord=np.argmax(data,axis=0)[0]
    return data[cord]

nlist=np.linspace(0.01,0.99,1000)

optdata=[]

for i in nlist:
    optdata.append(EMopt(i,1000,r2max))
    
Idata=np.transpose(optdata)[0]
rdata=np.transpose(optdata)[1]
r2data=np.transpose(optdata)[2]
```

```python

plt.title(r"Optimal $I_Q$ vs. $\eta$ of Mandel Interferometer With All Modes for $N=$"+str(N)+", $T=1$, $r_{2_{max}}=$"+str(r2max))
plt.xlabel(r"Transmitivity $\eta$")
plt.ylabel(r"Max $I_Q$")
plt.plot(nlist,Idata,label=r"Mandel with all modes")
plt.plot(nlist,MZI(nlist),color="black",linestyle="dashed",label="MZI")
plt.legend()
plt.show()

plt.title(r"Optimal Parameters vs. $\eta$ for $N=$"+str(N)+", $T=1, r_{2_{max}}=$"+str(r2max))
plt.xlabel(r"Transmitivity $\eta$")
plt.plot(nlist,rdata,label=r"Optimal $r_1$")
plt.plot(nlist,r2data,label=r"Optimal $r_2$",color="black")
plt.legend()
plt.show()

```
